```python
import matplotlib.pyplot as plt

results = locfdr(data)
print('Executed simulation described in Section 4 of the locfdr() R vignette at http://cran.r-project.org/web/packages/locfdr/vignettes/locfdr-example.pdf . See results in variable \'results\'.')
plt.show()
```

```python

import matplotlib.pyplot as plt
fig = plt.figure(figsize = (14, 8))

bins = np.linspace(min(data), max(data), 120)
plt.hist(data, bins = bins, fill=False)
poisson_density = PoissonDensity().fit(data)
plt.plot(bins, poisson_density.score_samples(bins), color='#3bbf53', linewidth = 3)
plt.xlim(min(data), max(data))
plt.show()
```
